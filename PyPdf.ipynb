```python
from pypdf import PdfReader
import re

reader = PdfReader("gs2023.pdf")
number_of_pages = len(reader.pages)

all_text = []

for i in range(number_of_pages):
    page = reader.pages[i]
    text = page.extract_text()
    all_text.append(text)

# Join the list of text into a single string
final_text = '\n'.join(all_text)
# Remove specific text occurrences
text_to_remove = ["Prepare with DrishtiIAS", "Classroom", "Courses", "SCAN MEOnline", "Courses", "SCAN ME SCAN MEBooks &", "MagazinesUPSC Mains Test", "Series", "2023 (GS/", "Optionals/Essay)", "", "", "SCAN ME"]

for text in text_to_remove:
    final_text = final_text.replace(text, "")
final_text = final_text.replace('\n','')
# Remove occurrences like I1, I2, I3, etc.
final_text = re.sub(r'I\d+', '', final_text)
final_text = re.sub(r'General Studies Paper-\d+', '', final_text)
final_text = re.sub(r'ISet-[A-Za-z]', '', final_text)

# Now 'final_text' contains the text from all pages as a single string
print(final_text)
```

```python
# Assuming 'text' is a multiline string

###final splitting text
# Remove newlines
text = final_text
text = text.replace("\n", "")
# text = re.sub(r'\b(b)\)(?![\w)])', r'(\1)', text)

# text = text.replace("b)","(b)")
# text = text.replace("a)","(a)")
# text = text.replace("c)","(c)")
# text = text.replace("d)","(d)")

# Split based on (a), (b), (c), (d)
result = re.split(r'\s*(?=\(\w\))', text)
# result = re.split(r'\s*(?=[\(a-d]\))', text)

# Combine each pair of substrings and splitting characters
pairs = [result[i] + (result[i + 1] if i + 1 < len(result) else '') for i in range(0, len(result), 2)]

# Iterate through pairs and split only if (d) is present and consider only the first number after (d)
final_result = []
for pair in pairs:
    if "(d)" in pair:
        # Find the index of (d)
        d_index = pair.find("(d)")
        if d_index != -1:
            # Extract the content after (d) including (d)
            content_after_d = pair[d_index:]
            # Split based on the first occurrence of a number
            subparts = re.split(r'(\d+\.)', content_after_d, 1)
            # Remove empty strings from the result
            subparts = [part.strip() for part in subparts if part.strip()]
            final_result.extend(subparts)
        else:
            final_result.append(pair)
    else:
        final_result.append(pair)

    # if "b)" in pair and "b)" not in pair:
    #     d_index = pair.find("b)")
    #     if d_index != -1:
    #       pair[d_index] = "(b)"
    #       final_result.append(pair)

# Print the final result
final_result = ['1.']+final_result
print(final_result)
```

```python
import pandas as pd
from google.colab import files

# Assuming you have an array named 'final_text_array' with the specified pattern
final_text_array = final_result

# Ensure all arrays have the same length by filling missing values with None
max_length = max(len(final_text_array[0::6]), len(final_text_array[1::6]), len(final_text_array[2::6]),
                 len(final_text_array[3::6]), len(final_text_array[4::6]), len(final_text_array[5::6]))

final_text_array += [None] * (max_length * 6 - len(final_text_array))

# Create a DataFrame
data = {
    'Question_Number': final_text_array[0::6],
    'Question': final_text_array[1::6],
    'Option_1': final_text_array[2::6],
    'Option_2': final_text_array[3::6],
    'Option_3': final_text_array[4::6],
    'Option_4': final_text_array[5::6],
    'Paper': "UPSC Civil Services (Preliminary) Exam GS-I",
    'Year':"2023",
    'Set':"Set-B"
}

df = pd.DataFrame(data)

# Export DataFrame to Excel
df.to_excel('output2023.xlsx', index=False)
files.download('output2023.xlsx')
```
